```python
from sklearn.decomposition import IncrementalPCA    # inital reduction
from sklearn.manifold import TSNE                   # final reduction
import numpy as np                                  # array handling
import os
import gensim
import pandas as pd
```

```python
os.chdir(r'C:\Users\julie\Dropbox\Julien\python\asr_framework\src')
model=gensim.models.KeyedVectors.load('../open-data/modeles_gensim/asr_global_wv')
```

```python
def reduce_dimensions(model, three_dim=False,perplexity=30,n_iter=1000):
    num_dimensions = 3 if three_dim else 2  # final num dimensions (2D, 3D, etc)

    # extract the words & their vectors, as numpy arrays
    vectors = np.asarray(model.wv.vectors)
    labels = np.asarray(model.wv.index_to_key)  # fixed-width numpy strings

    # reduce using t-SNE
    tsne = TSNE(n_components=num_dimensions, random_state=0,init='pca',learning_rate='auto',n_jobs=-1,perplexity=perplexity,n_iter=n_iter)
    vectors = tsne.fit_transform(vectors)

    x_vals = [v[0] for v in vectors]
    y_vals = [v[1] for v in vectors]
    if three_dim:
        z_vals = [v[2] for v in vectors]
        return x_vals, y_vals, z_vals, labels
    else:
        return x_vals, y_vals, labels
```

```python
x_vals, y_vals, z_vals, labels = reduce_dimensions(model,three_dim=True,perplexity=5,n_iter=10000)
```

```python
df = pd.DataFrame(zip(x_vals, y_vals, z_vals, labels),columns=['x', 'y', 'z','label'])
df.to_csv('asr-themes-sample_3d-wlabels.csv',sep=',',header=True,encoding='utf-8',index=False)
```

```python
def reduce_to_one_dimension(wv_model, perplexity=30,n_iter=1000):
    # extract the words & their vectors, as numpy arrays
    vectors = np.asarray(wv_model.vectors)
    labels = np.asarray(wv_model.index_to_key)  # fixed-width numpy strings

    # reduce using t-SNE
    tsne = TSNE(n_components=1, random_state=0,init='pca',learning_rate='auto',n_jobs=-1,perplexity=perplexity,n_iter=n_iter)
    vectors = tsne.fit_transform(vectors)

    x_vals = [v[0] for v in vectors]
    return x_vals, labels
```

```python
x_vals, labels = reduce_to_one_dimension(model)
```

```python
df = pd.DataFrame(zip(x_vals, labels),columns=['x_vals','labels'])
df=df.sort_values(by=['x_vals'],ascending=True)
df.to_csv('one_dim_wlabels.csv',sep='\t',header=False,encoding='ansi',index=False)
```
